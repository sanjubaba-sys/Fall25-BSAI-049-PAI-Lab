**OpenCV Image Processing**


```python
# Step 0: Import Libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to display image nicely
def show_image(img, title="Image", cmap=None):
    plt.figure(figsize=(6,6))
    plt.imshow(img, cmap=cmap)
    plt.title(title)
    plt.axis("off")
    plt.show()

# Load image (replace 'image.jpg' with your file)
img = cv2.imread("image.jpg")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
show_image(img_rgb, "Original Image")
```

**Blurring an Image**

```python
blur = cv2.GaussianBlur(img_rgb,(7,7),0)
show_image(blur, "Blurred Image")
```

**Grayscaling of Images**

```python
gray = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)
show_image(gray, "Grayscale Image", cmap="gray")
```

**Scaling, Rotating, Shifting and Edge Detection**

```python
scaled = cv2.resize(img_rgb, (400,300))
show_image(scaled, "Scaled Image")
```

```python
(h,w) = img_rgb.shape[:2]
center = (w//2, h//2)
M_rot = cv2.getRotationMatrix2D(center, 45, 1.0)
rotated = cv2.warpAffine(img_rgb, M_rot, (w,h))
show_image(rotated, "Rotated Image")
```

```python
M_shift = np.float32([[1,0,100],[0,1,50]])
shifted = cv2.warpAffine(img_rgb, M_shift, (w,h))
show_image(shifted, "Shifted Image")
```

```python
edges = cv2.Canny(gray,100,200)
show_image(edges, "Edge Detection", cmap="gray")
```

**Intensity Transformation (Brightness/Contrast)**

```python
bright = cv2.convertScaleAbs(img_rgb, alpha=1.5, beta=50)
show_image(bright, "Brightness Increased")
```

**Image Translation (similar to shifting, shown above)**

```python
tx, ty = 50, 100
M_trans = np.float32([[1,0,tx],[0,1,ty]])
translated = cv2.warpAffine(img_rgb, M_trans, (w,h))
show_image(translated, "Translated Image")
```

**Image Pyramid**

```python
smaller = cv2.pyrDown(img_rgb)
show_image(smaller, "Pyramid Down")

larger = cv2.pyrUp(smaller)
show_image(larger, "Pyramid Up")
```

**Histogram Equalization**

```python
gray = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)
equalized = cv2.equalizeHist(gray)
show_image(equalized, "Histogram Equalized", cmap="gray")
```

**Convert Image to Another Color Space**

```python
hsv = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2HSV)
show_image(hsv, "HSV Color Space")
```

**Visualizing Image in Different Color Spaces**

```python
# RGB
show_image(img_rgb, "RGB Image")

# HSV
show_image(hsv, "HSV Image")

# Grayscale
show_image(gray, "Grayscale Image", cmap="gray")
```

**Create Border Around Image**

```python
border = cv2.copyMakeBorder(img_rgb, 50,50,50,50, cv2.BORDER_CONSTANT, value=[255,0,0])
show_image(border, "Image with Border")
```

```python
# Water Jug Problem using DFS (Detailed Python Code)

# Capacity of the jugs
JUG1_CAPACITY = 4
JUG2_CAPACITY = 3

# Target amount
TARGET = 2

# To keep track of visited states
visited_states = set()


def print_state(jug1, jug2, rule):
    print("Rule Applied:", rule)
    print("Current State -> Jug1:", jug1, " Jug2:", jug2)
    print("----------------------------------")


def dfs(jug1, jug2):
    
    # If we have already visited this state, stop
    if (jug1, jug2) in visited_states:
        return False
    
    visited_states.add((jug1, jug2))
    
    # If goal is reached
    if jug1 == TARGET or jug2 == TARGET:
        print("Goal Achieved!")
        print("Final State -> Jug1:", jug1, " Jug2:", jug2)
        return True

    # Rule 1: Fill Jug1
    new_jug1 = JUG1_CAPACITY
    new_jug2 = jug2
    print_state(new_jug1, new_jug2, "Fill Jug1")
    if dfs(new_jug1, new_jug2):
        return True

    # Rule 2: Fill Jug2
    new_jug1 = jug1
    new_jug2 = JUG2_CAPACITY
    print_state(new_jug1, new_jug2, "Fill Jug2")
    if dfs(new_jug1, new_jug2):
        return True

    # Rule 3: Empty Jug1
    new_jug1 = 0
    new_jug2 = jug2
    print_state(new_jug1, new_jug2, "Empty Jug1")
    if dfs(new_jug1, new_jug2):
        return True

    # Rule 4: Empty Jug2
    new_jug1 = jug1
    new_jug2 = 0
    print_state(new_jug1, new_jug2, "Empty Jug2")
    if dfs(new_jug1, new_jug2):
        return True

    # Rule 5: Pour Jug1 -> Jug2
    transfer = min(jug1, JUG2_CAPACITY - jug2)
    new_jug1 = jug1 - transfer
    new_jug2 = jug2 + transfer
    print_state(new_jug1, new_jug2, "Pour Jug1 to Jug2")
    if dfs(new_jug1, new_jug2):
        return True

    # Rule 6: Pour Jug2 -> Jug1
    transfer = min(jug2, JUG1_CAPACITY - jug1)
    new_jug1 = jug1 + transfer
    new_jug2 = jug2 - transfer
    print_state(new_jug1, new_jug2, "Pour Jug2 to Jug1")
    if dfs(new_jug1, new_jug2):
        return True

    return False


def start_water_jug_problem():
    print("Starting Water Jug Problem Using DFS")
    print("Jug1 Capacity:", JUG1_CAPACITY)
    print("Jug2 Capacity:", JUG2_CAPACITY)
    print("Target:", TARGET)
    print("----------------------------------")
    
    dfs(0, 0)


# Run the program
start_water_jug_problem()
```
